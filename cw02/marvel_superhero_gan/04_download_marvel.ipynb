```python
import os
import requests
import urllib
import hashlib
import time
```

```python
public_key = ''
private_key = ''
```

```python
url = "https://gateway.marvel.com:443/v1/public/characters"

params = {
    "limit": 100, 
    "offset": 0, 
    "apikey": public_key,
    "hash": ''
}
```

```python
def get_request_hash(private_key, public_key):
    ts = str(time.time())
    m = hashlib.md5()
    m.update(ts.encode('utf-8'))
    m.update(private_key.encode('utf-8'))
    m.update(public_key.encode('utf-8'))
    return ts, m.hexdigest()
```

```python
# Set the download directory
download_dir = "data/marvel_thumbnails/images"
if not os.path.exists(download_dir):
    os.makedirs(download_dir)
```

```python
# Send the first request to get the total count of characters
params['ts'], params['hash'] = get_request_hash(private_key, public_key)

response = requests.get(url, params=params).json()
total_count = response["data"]["total"]
```

```python
# Loop through all characters and download their thumbnails
for offset in range(0, total_count, 100):
    # Update the offset parameter in the query params
    params['ts'], params['hash'] = get_request_hash(private_key, public_key)
    params["offset"] = offset
    
    # Send the request to get the characters
    response = requests.get(url, params=params).json()
    
    # Loop through each character and download its thumbnail
    for character in response["data"]["results"]:
        # Get the thumbnail URL
        thumbnail_url = character["thumbnail"]["path"] + "." + character["thumbnail"]["extension"]
        
        # Set the filename to save the thumbnail
        filename = os.path.join(download_dir, character["name"] + ".jpg")
        
        # Download the thumbnail and save it to file
        try:
            response = requests.get(thumbnail_url)
            with open(filename, "wb") as f:
                f.write(response.content)
                
            print(f"Downloaded thumbnail for {character['name']}")
        except:
            print(f"Skipping {filename} - file not found.")
```
